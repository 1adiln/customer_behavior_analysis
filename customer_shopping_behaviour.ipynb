```python
# Import Libraries
import pandas as pd
```

```python
# Import Dataset
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Converting Column names into snake case
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
# Checking column names again
df.columns
```

```python
# Renaming 'purchase_amount_(usd)' column name
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df.columns
```

#### Feature Engineering

```python
# 1.Create a column for Age Group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)

# what is pd.qcut?
```

```python
df[[ 'age', 'age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# 2.Create a column for purchase frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df.isnull().sum()
```

```python
# Converting Data Type
df['purchase_frequency_days'] = df['purchase_frequency_days'].astype(int)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
# Checking Discount vs Promo code Columns
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
# To exactly check if they totally same or not
(df['discount_applied'] == df['promo_code_used']).all() 
```

```python
df = df.drop('promo_code_used', axis=1)
```

```python
df.columns
```

```python
df.to_csv("customer_shopping_behavior_cleaned.csv", index=False)
```
